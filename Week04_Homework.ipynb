```python
import numpy as np
import pandas as pd 
from sklearn import datasets
from sklearn import linear_model
```

<h3>Simple Regression Using statsmodel</h3>

```python
boston = datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df.head()
```

```python
df.describe()
```

```python
#Assigning a different dataframe for housing value-'MEDV' and printing the first 5 rows
target = pd.DataFrame(boston.target, columns=["MEDV"])
```

```python
target.head()
```

```python
import statsmodels.api as sm
#Define 'RM' as X or the feature and 'MEDV' as y or the label
X = df["RM"]
y = target["MEDV"]


result = sm.OLS(y, X).fit()
prediction = result.predict(X) 

# Printing the statistics
result.summary()
```

<h3>Multiple Regression Using statsmodel </h3>                                    

```python
X = df[['RM', 'LSTAT']]
y = target['MEDV']
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

<h3>What does a coefficient of determination of R2= .948 mean? </h3>

 R squared value of 0.948 means that the model explains 94.8% of variance in the dependent variable.

<h3>Is there any autocorrelation? If not, why? </h3>

According to durbin Watson Test, if a autocorrelation is in between 0 and 2 the autocorrelation is positive. 
In this case, the durbin watson test value is 0.834 which is in between 0 and 2.

<h3>Are the data coming from a normal distribution? Why? </h3>

Through the inference of OLS we got the skew value(1.351) which is positive, that means the distribution is right skewed and not normally distributed.

<h3>Multiple Regression Using scikitlearn</h3>

```python
X = df[['RM', 'LSTAT']]
y = target['MEDV']
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X,y)
```

```python
predictions = lm.predict(X)
print(predictions)
```

```python
#R^2 value
lm.score(X,y)
```

```python
#printing the coefficients
lm.coef_
```
