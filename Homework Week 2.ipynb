```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mu, sigma = 3, 2 
s = np.random.normal(mu, sigma, 10000)
mu, std = norm.fit(s)
```

```python
plt.hist(s, bins=25, density=True, alpha=0.8)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p,'k', linewidth=2,color="r")
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```

```python
sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean, sample_sigma, 200)
mu, std = norm.fit(sample)
```

```python
plt.hist(sample, bins=25, density=True, alpha=0.8)
xmin1, xmax1 = plt.xlim()
x1 = np.linspace(xmin1, xmax1, 100)
p1 = norm.pdf(x1, sample_mean, std)
plt.plot(x1, p1, 'k', linewidth=2,color="r")
title = "Fit results: sample_mean = %.2f  sample_sigma = %.2f" % (sample_mean, sample_sigma)
plt.title(title)

plt.show()
```

```python
plt.hist(s, bins=30, alpha=0.1, label='x',density=True)
plt.plot(x1, p1, 'k', linewidth=2,color="r")
plt.hist(sample, bins=30, alpha=0.1, label='y',density=True)
plt.plot(x, p, 'k', linewidth=2,color="b")
plt.show()
```
